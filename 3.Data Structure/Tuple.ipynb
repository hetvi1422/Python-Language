```python
numbers=(1,2,3,3,4)
print(numbers.count(3))
print(type(numbers))
print(numbers.index(4))
```

```python
##packing of tuple

packed_tuple=1,"hey",2.5
print(packed_tuple)

##unpacking if tuple
a,b,c=packed_tuple

print(a)
print(b)
print(c)
```

```python
##unpacking with*

n=(1,2,3,4,5,6)
first,middle,*last=n
print(first)
print(middle)
print(*last)
```

```python
##nested list
lst=[[1,2,3],[4,5,6],[7,8,9]]
print(lst[2][0:])

##tuple in list
lst=[[1,2,3],[4,5,6],("hey",1,2)]
print(type(lst[2]))

##nested tuple
tpl=((1,2,3),("hey",3.14),(True,False))
print(tpl[1])
```

```python
##iterating in tuple

for sub_tuple in tpl:
    for item in sub_tuple:
        print(item,end=" ")
    print()
    
```
