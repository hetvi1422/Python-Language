```python
my_set={1,2,3,4,5,6,5,6,7}
print(my_set)
```

```python
##add or remove operation in set

my_set.add(8)
print(my_set)
my_set.remove(5)
print(my_set)
```

```python
##remove element
my_set.remove(10)
print(my_set)
## When element is not present in your set an dremove it so it will give you error.
```

```python
##discard operation
my_set.discard(10)
print(my_set)
'''when you Discard use if element is not present in your set so it will not show it  but actual set is give in
 output and error is not come.'''
```

```python
##pop operation
removed_element=my_set.pop
print(removed_element)
print(my_set)
```

```python
##clear set

my_set.clear()
print(my_set)
```

```python
##Membership test in set
my_set={1,2,3,4}
print(3 in my_set)
print(10 in my_set)
```

```python
##Mathematical operation

set1={1,2,3,4,5}
set2={3,4,5,6,7}

set_union=set1.union(set2)
set_intersection=set1.intersection(set2)
set_intersection_update=set1.intersection_update(set2)
set_difference=set1.difference(set2)
set_symmetric_difference=set1.symmetric_difference(set2)

print(set_union)
print(set_intersection)
print(set_intersection_update)
print(set_difference)
print(set_symmetric_difference)


```

```python
##Counting unique words in text
Text="today is sunday."
words=Text.split()

##convert list of  words to set to get unique word
unique_word=set(words)
print(unique_word)
print(len(unique_word))
```
